```python
def ChickenRabbit(heads,legs):
    rabbits=legs/2-heads
    chicken=heads-rabbits
    s="There are "+str(chicken)+" chicken and "+str(rabbits)+" rabbits"
    return s
ChickenRabbit(35,94)
```

```python
import os
import fnmatch
from collections import deque
def get(path):

        queue = deque()
        queue.append(path)
        while len(queue)>0:
            temp = queue.popleft()
            try:
                list = os.listdir(temp)
            except WindowsError as e:
                pass
            for filename in list:
                next = os.path.join(temp,filename)
                if os.path.isfile(next):
                    if fnmatch.fnmatch(filename,'*.dll'):
                        print (filename)
                elif os.path.isdir(next):
                    queue.append(next)

get('C:\\')
```
